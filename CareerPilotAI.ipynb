## Resume Agent

The Resume Agent analyzes resumes and provides:

- Resume Score
- Strengths
- Weaknesses
- Suggested Improvements

```python
def resume_agent(resume_text):

    prompt = f"""
    You are an expert resume reviewer.

    Analyze the following resume and provide:

    1. Resume Score (out of 100)
    2. Strengths
    3. Weaknesses
    4. Suggested Improvements

    Resume:
    {resume_text}
    """

    response = model.generate_content(prompt)

    return response.text
```

```python
sample_resume = """
B.Tech Data Science Student

Skills:
Python
SQL
Selenium

Projects:
Used Car Price Prediction
New York Airbnb Analytics

Education:
SRM University AP
"""

print(resume_agent(sample_resume))
```

## Interview Agent

The Interview Agent generates beginner, intermediate, and advanced interview questions with answers for a target role.

```python
def interview_agent(role):

    prompt = f"""
    You are a technical interviewer.

    Generate:
    5 beginner questions
    5 intermediate questions
    5 advanced questions

    for the role: {role}

    Also provide answers.
    """

    response = model.generate_content(prompt)

    return response.text
```

## Skill Gap Agent

The Skill Gap Agent compares a student's current skills with a target role and identifies:

- Matching Skills
- Missing Skills
- Recommended Learning Path

```python
def skill_gap_agent(current_skills, target_role):

    prompt = f"""
    Compare the student's skills with the target role.

    Current Skills:
    {current_skills}

    Target Role:
    {target_role}

    Provide:
    1. Matching skills
    2. Missing skills
    3. Recommended learning path
    """

    response = model.generate_content(prompt)

    return response.text
```

```python
def resume_agent(resume_text: str) -> str:
    """Analyzes a resume and provides structured feedback."""
    prompt = f"""
    You are an expert resume reviewer.
    Analyze the following resume and provide:
    1. Resume Score (out of 100)
    2. Strengths
    3. Weaknesses
    4. Suggested Improvements

    Resume:
    {resume_text}
    """
    response = model.generate_content(contents=prompt)
    return response.text

def interview_agent(role: str) -> str:
    """Generates a structured list of technical interview questions and answers."""
    prompt = f"""
    You are a technical interviewer.
    Generate:
    - 5 beginner questions
    - 5 intermediate questions
    - 5 advanced questions
    for the role: {role}
    Also provide comprehensive answers for each.
    """
    response = model.generate_content(contents=prompt)
    return response.text

def skill_gap_agent(resume_text: str, target_role: str) -> str:
    """Compares a resume against a target role to find missing skills."""
    prompt = f"""
    You are a Career Transition Specialist.
    Compare the following resume against the industry requirements for a '{target_role}' role.
    Identify:
    1. Matching skills the candidate already possesses.
    2. Critical skill gaps (Technical, Tools, and Soft Skills) they need to bridge.

    Resume:
    {resume_text}
    """
    response = model.generate_content(contents=prompt)
    return response.text

def roadmap_agent(skill_gaps: str, target_role: str) -> str:
    """Creates a step-by-step learning schedule to bridge gaps."""
    prompt = f"""
    You are an expert Technical Career Coach.
    Based on the following identified skill gaps for someone targeting a '{target_role}' role,
    generate a detailed 4-week structured learning roadmap to bridge these gaps.

    Skill Gaps/Context:
    {skill_gaps}
    """
    response = model.generate_content(contents=prompt)
    return response.text
```

# Career Agent (Orchestrator)

The CareerAgent acts as the central coordinator of CareerPilot AI.

It performs:

1. Resume Evaluation
2. Skill Gap Analysis
3. Learning Roadmap Generation
4. Interview Preparation

The outputs from all specialized agents are combined into a single career guidance report.

```python
def career_agent(resume_text: str, target_role: str) -> str:
    """
    Main Orchestrator Agent. Coordinates the resume reviewer, skill gap analyzer,
    and roadmap generator to output a comprehensive career guidance package.
    """
    print("⏳ Step 1: Evaluating your resume...")
    resume_feedback = resume_agent(resume_text)

    print("⏳ Step 2: Analyzing skill gaps for your target role...")
    gap_analysis = skill_gap_agent(resume_text, target_role)

    print("⏳ Step 3: Generating your personalized preparation roadmap...")
    learning_roadmap = roadmap_agent(gap_analysis, target_role)

    print("⏳ Step 4: Curating sample interview prep material...")
    interview_prep = interview_agent(target_role)

    # Compile the final report
    final_report = f"""
# =======================================================
#   🚀 CAREER PILOT AI REPORT: {target_role.upper()}
# =======================================================

## 📋 RESUME EVALUATION & SCORE
{resume_feedback}

---

## 🔍 SKILL GAP ANALYSIS
{gap_analysis}

---

## 🗺️ YOUR 4-WEEK PREPARATION ROADMAP
{learning_roadmap}

---

## 💼 PRACTICE INTERVIEW QUESTIONS
{interview_prep}
    """
    return final_report
```
